```python
import transformers

print(transformers.__version__) ## 새 주석: transformer 버전 확인
```

```python
from transformers import pipeline

senti = pipeline('sentiment-analysis') ## 새 주석: 감성 분석 모듈 호출
```

```python
senti.model ## 새 주석: 감성 분석에 쓰이는 모델 정보
```

```python
text_1 = "A few days ago, I bought this product. It is sooooo good that I do not use it! It is too useful for me!"
## 새 주석: 예제문 변경, 어떤 상품에 대한 비꼬는 리뷰.
print(senti([text_1]))
```

```python
text_2 = "What? You accidentally deleted our final report? Oh, this is just great. Now we have to do it all over again."
## 새 주석: 예제문 변경, 최종 보고서를 삭제해서 처음부터 다시 해야하는 상황을 비꼬는 상황.
print(senti([text_2]))
```
